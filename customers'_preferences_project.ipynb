#**Outlier Detection using IQR Method**

 ----------------------------------------
 **Objective :** Detect outliers in each numeric column using the IQR method and count the number of outliers per column and in total rows




```python
#  Required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Select numeric columns only
numeric_cols = df.select_dtypes(include=['number']).columns

 #Calculate IQR for each numeric column
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
  #  Detect outliers: values outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
outlier_mask = (df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))

# Count outliers per column
outliers_per_column = outlier_mask.sum()

#  Count rows that contain at least one outlier
rows_with_outliers = outlier_mask.any(axis=1).sum()

# Display results
print(" Number of outliers per numeric column:")
print(outliers_per_column)

print(f"\n Total rows with at least one outlier: {rows_with_outliers}")

#Show a boxplot for visualization
plt.figure(figsize=(15, 8))
df[numeric_cols].boxplot()
plt.xticks(rotation=45)
plt.title("Boxplot for Numeric Columns")
plt.grid(True)
plt.show()
```

```python
# Remove rows that contain outliers
df_clean = df[~outlier_mask.any(axis=1)]

# Show new shape of the data
print("\n Original shape:", df.shape)
print(" Cleaned shape (after removing outliers):", df_clean.shape)

# Reset index if needed
df_clean.reset_index(inplace=True)

#  Preview cleaned data
df_clean.head()
```


#  **Application of Mining Technique – K-Medoids Clustering**

Objective:
* To group similar stores based on their numeric performance features using the K-Medoids clustering algorithm. This helps the business identify patterns in store performance and customer preferences.

---

###  Technique Used: K-Medoids Clustering

- **Reason for choice:**  
  K-Medoids is more robust to outliers compared to K-Means. Since our dataset had outliers (which we previously removed), K-Medoids remains a reliable clustering technique for store segmentation.

---

###  Steps Taken:

1. **Selected Features:**
   - Only numerical columns were used for clustering.
   - The target variable `Basket:2014` was excluded from the clustering process.

2. **Normalization:**
   - Min-Max normalization was applied to bring all feature values into the [0, 1] range.
   - This ensures fair distance calculation between features during clustering.

3. **K-Medoids Application:**
   - Chose number of clusters `k = 3`.
   - Trained the K-Medoids algorithm and assigned each store to one cluster.

4. **Result Analysis:**
   - Added the resulting `Cluster` label to the original data.
   - Analyzed average values per cluster to interpret their characteristics.

5. **Visualization:**
   - A pairplot was generated to visualize how data points from different clusters relate across features.

---

Output Summary:
* Each store was classified into one of three clusters based on performance metrics such as Sales, Profit, Labor Hours, and Gross Margin. This classification helps in better customer targeting and strategic planning.

```python
#  1. Import required libraries
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.utils import calculate_distance_matrix
import random
```

```python
#  Select only numeric features for clustering
features = df_clean.select_dtypes(include=['float64', 'int64'])
```

```python
#  Compute distance matrix for K-Medoids
distance_matrix = calculate_distance_matrix(features.values)
```

```python
#  Choose the number of clusters and initialize medoids
k = 3  # You can change to 2, 4, etc.
initial_medoids = random.sample(range(len(features)), k)
```

```python
#  Apply K-Medoids algorithm
kmedoids_instance = kmedoids(distance_matrix, initial_medoids, data_type='distance_matrix')
kmedoids_instance.process()
```

```python
#  Get clusters and final medoids
clusters = kmedoids_instance.get_clusters()
final_medoids = kmedoids_instance.get_medoids()
```

```python
print("Final Medoids Indexes:", final_medoids)
print("Cluster Assignments:")
for i, cluster in enumerate(clusters):
    print(f"Cluster {i+1} -> {len(cluster)} stores")
```

```python
#  Add cluster label to the DataFrame
cluster_labels = [0] * len(features)
for cluster_index, cluster in enumerate(clusters):
    for index in cluster:
        cluster_labels[index] = cluster_index

df_clean['Cluster'] = cluster_labels
```

```python
#  Show sample of the clustered data
print(df_clean[['Cluster'] + list(features.columns)].head())
```

# **Import Libraries for Hierarchical**
* importing essential libraries for data manipulation `pandas, numpy`, scaling `StandardScaler`, clustering, evaluation `silhouette_score`, and visualization `matplotlib, seaborn`.

```python
# Import necessary libraries for Hierarchical
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Feature Selection**
* select only numerical features from your cleaned dataset **`df_clean`** for clustering. Clustering doesn’t work directly on categorical data.

# **Data Scaling**
* **`Scaling`** ensures that all features contribute equally to the distance calculations.

* **`StandardScaler`** standardizes data to have mean = 0 and standard deviation = 1.

```python
# Select the features you want to use for clustering
features = df_clean.select_dtypes(include=['float64', 'int64'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

# **Create Clustering Models with Different Linkage Methods**
* Create four **hierarchical** clustering models with different linkage criteria:

 * **Single:** Minimum distance between points in clusters.

 * **Complete**: Maximum distance between points in clusters.

 * **Average**: Average distance between points in clusters.

 * **Ward**: Minimizes the variance within each cluster (best for continuous data).


---



# **Fit the Models and Get Cluster Labels**
* Each model is applied to the scaled data, and you get the cluster labels (i.e., which cluster each data point belongs to).

```python
# Create Clustering Models with Different Linkage Methods
clusterer1 = AgglomerativeClustering(n_clusters=4, linkage='single')
clusterer2 = AgglomerativeClustering(n_clusters=4, linkage='complete')
clusterer3 = AgglomerativeClustering(n_clusters=4, linkage='average')
clusterer4 = AgglomerativeClustering(n_clusters=4, linkage='ward')
# You can adjust n_clusters, affinity, and linkage parameters

# Fit the model to the data
clusterer1.fit_predict(X_scaled)
clusterer2.fit_predict(X_scaled)
clusterer3.fit_predict(X_scaled)
clusterer4.fit_predict(X_scaled)

# Get the cluster labels
labels1 = clusterer1.labels_
labels2 = clusterer2.labels_
labels3 = clusterer3.labels_
labels4 = clusterer4.labels_
```

# **Evaluate Clustering Quality with Silhouette Score**
* compute the Silhouette Score for each clustering result:

  * Score near 1 → well-separated clusters

  * Score near 0 → overlapping clusters

  * Score below 0 → poor clustering (some points may be in the wrong cluster)

```python
print("Silhouette Score (Single):", silhouette_score(X_scaled, labels1))
print("Silhouette Score (Complete):", silhouette_score(X_scaled, labels2))
print("Silhouette Score (Average):", silhouette_score(X_scaled, labels3))
print("Silhouette Score (Ward):", silhouette_score(X_scaled, labels4))
```

```python
scores = [
    silhouette_score(X_scaled, labels1),
    silhouette_score(X_scaled, labels2),
    silhouette_score(X_scaled, labels3),
    silhouette_score(X_scaled, labels4)
]
methods = ['Single', 'Complete', 'Average', 'Ward']

plt.figure(figsize=(6, 4))
sns.barplot(x=methods, y=scores)
plt.ylabel('Silhouette Score')
plt.title('Comparison of Linkage Methods')
plt.show()
```

# **Dendrogram Visualization**
* **`use linkage()`** to compute the linkage matrix required for plotting dendrograms (tree-like structures showing cluster merging).

* **`dendrogram()`** plots the hierarchical structure so you can visually inspect the cluster relationships.

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Calculate the linkage matrix
linked1 = linkage(X_scaled, 'single')
linked2 = linkage(X_scaled, 'complete')
linked3 = linkage(X_scaled, 'ward')
linked4 = linkage(X_scaled, 'average')

# Plot the dendrogram
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1),dendrogram(linked1, distance_sort='descending',show_leaf_counts=True),plt.title('Single')
plt.subplot(2, 2, 2),dendrogram(linked2, distance_sort='descending', show_leaf_counts=True),plt.title('Complete')
plt.subplot(2, 2, 3),dendrogram(linked3, distance_sort='descending', show_leaf_counts=True),plt.title('Ward')
plt.subplot(2, 2, 4),dendrogram(linked4, distance_sort='descending', show_leaf_counts=True),plt.title('Average')
plt.show()
```

Generally, a higher Silhouette Score indicates better-defined clusters. in this case:

* The **Ward** method seems to be performing the best, with the highest Silhouette Score of **0.1089**. This suggests that Ward's method is generating the most cohesive clusters compared to the other methods.

* **Single, Complete,** and **Average** methods are yielding lower scores, which could indicate that these clustering approaches are less effective for the dataset or the way they handle distances and merging clusters.
